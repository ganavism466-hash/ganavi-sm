```python
def linear_search(list1, n, key):
    for i in range(0, n):
        if list1[i] == key:
            return i
    return -1
list1=[1,3,4,5,7,9]
key=7
n = len(list1)
res=linear_search(list1, n, key)

if res ==-1:
     print("element not found")
else:
     print("element found at index:",res)
        
```

```python
graph = {
    '5':['3','7'],
    '3':['2','4'],
    '2':[],
    '7':['8'],
    '8':[]
}
visited=[]
queue=[]

def bfs (node):
    visited.append(node)
    queue.append(node)

    while queue:
        m = queue.append(node)
        print(m,end="")

        for nieghbour in graph[m]:
            if neighbour not in visited:
                visited.append(nieghbour)
                queue.append(neighbour)
        print("Breadth First Search:")
        bfs('s')
    
```

```python
graph = {
    '5':['3','7'],
    '3':['2','4'],
    '2':[],
    '7':['8'],
    '8':[]
}
visited = set()

def dfs(visited,graph,node):
    if node not in visited:
        print (node)
        visited.add(node)

        for neighbour in graph[node]:
            bfs(visited, graph,neighbour)

print("following in depth-first search:")
dfs (visited ,graph,'s')
        
```
